Correlation heatmap

```python
df_encoded = df.copy()

# Convert Attrition to binary 
df_encoded['Attrition'] = df_encoded['Attrition'].map({'Yes': 1, 'No': 0})

# Encode key categorical features for correlation
from sklearn.preprocessing import LabelEncoder

# Select top features to encode 
cols_to_encode = ['BusinessTravel', 'Department', 'JobRole', 'Gender', 'MaritalStatus', 'EducationField']

le = LabelEncoder()
for col in cols_to_encode:
    df_encoded[col] = le.fit_transform(df_encoded[col])

# Now plot the updated correlation heatmap
plt.figure(figsize=(20, 15))
sns.heatmap(df_encoded.corr(), annot=True, fmt=".2f", cmap='coolwarm')
plt.title('🔗 Correlation Heatmap (Encoded)')
plt.show()
```

Age distribution

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x='Attrition', y='Age', data=df)
plt.title('Age vs Attrition')
plt.show()
```

Explore key categorical features

```python
categorical_features = ['Department', 'Gender', 'JobRole', 'EducationField', 'MaritalStatus']
for col in categorical_features:
    plt.figure(figsize=(8, 4))
    sns.countplot(data=df, x=col, hue='Attrition')
    plt.title(f'{col} Distribution by Attrition')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```
